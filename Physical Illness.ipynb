```python
# First importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("illness_percentage_by_job.csv")
df
```

```python
# Create figure first
plt.figure(figsize=(16, 10))

# X positions for bars
x_positions = np.arange(len(df))

# Draw bars
plt.bar(
    x_positions, 
    df['Percentage'], 
    edgecolor='black'
)

# Define fixed 5% error
errors = [np.std(df.Percentage)] * len(df)

# Add error bars
plt.errorbar(
    x_positions, 
    df['Percentage'], 
    yerr=errors, 
    fmt='none',       # no markers, just error bars
    ecolor='black', 
    capsize=4
)

# Formatting
plt.xticks(x_positions, df['Job'], rotation=90)
plt.xlabel("Job", fontsize=14)
plt.ylabel('Percentage', fontsize=14)
plt.title("Percentage of Employees Experiencing Physical Health Issues", fontsize=20)
plt.ylim(80, 100)

plt.show()
```

```python
data = pd.read_csv("illness_percentage_by_age.csv")
x = data['Age'].to_numpy()
y = data['Percentage'].to_numpy()

r = np.corrcoef(x, y)[0, 1]  # Pearson correlation
r2 = r**2
print("R:", r2)
```

```python
df = pd.read_csv("illness_percentage_by_hours.csv")
df
```

```python
plt.xlim(20,70)
plt.ylim(80,100)
plt.xlabel("Hours Per Week", size=14)
plt.ylabel("Percentage", size=14)
plt.title("Prevalence of Physical Illness by Hours Worked Per Week")
plt.style.use('default')
plt.scatter(df.Hours, df.Percentage, edgecolors='black')
```
