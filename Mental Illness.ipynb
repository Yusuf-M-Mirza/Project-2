```python
# First importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from scipy.stats import pearsonr
```

```python
# Obtain data from .csv file and store it as a DataFrame
df = pd.read_csv("mentalillness_by_region.csv")
df
```

```python
# Displays the standard deviation of Percentage
print(np.std(df.Percentage))
```

```python

# Plots a bar graph of the six regions against Percentage [of employees suffering mental illness] with standard error bars
x_positions = np.arange(len(df))

plt.bar(df.Region, df.Percentage, edgecolor='black')

standard_deviation = np.std(df.Percentage)

errors = [standard_deviation] * len(df)

plt.errorbar(
    x_positions,
    df['Percentage'],
    yerr=errors,
    fmt='none',
    ecolor='black',
    capsize=4
)

plt.ylabel("Percentage", size=12)
plt.title("Percentage of Employees Experiencing Mental Illness by Region", size=12)
plt.tick_params(labelrotation=45, axis='x')
plt.show()
```

```python
# Obtain data from .csv file and store it as a DataFrame
df = pd.read_csv("mentalillness_by_hours.csv")
df
```

```python
# Plot two line graphs on the same axis
plt.plot(df.Hours, df.Work_Life, linewidth=2.5, label="Work-Life Balance")
plt.plot(df.Hours, df.Social_Isolation, linewidth=2.5, label="Social Isolation")
plt.ylabel("Score", size=12)
plt.xlabel("Hours Per Week", size=12)
plt.xlim(35,65)
plt.legend()
plt.title("Mean Score by Hours Per Week")
plt.show()
```

```python
# Find the correlation coefficient between Hours and Work_Life
df = pd.read_csv("mentalillness_by_hours.csv")
x = df['Hours'].to_numpy()
y = df['Work_Life'].to_numpy()

r = np.corrcoef(x, y)[0, 1]
r2 = r**2
print("R²:", r2)
```

```python
# Find the correlation coefficient between Hours and Social_Isolation
df = pd.read_csv("mentalillness_by_hours.csv")
x = df['Hours'].to_numpy()
y = df['Social_Isolation'].to_numpy()

r = np.corrcoef(x, y)[0, 1]
r2 = r**2
print("R²:", r2)
```

```python
# We use Linear Regression to find the link between Work_Life and Social_Isolation
X = (1 / df["Work_Life"]).values.reshape(-1, 1)
y = df["Social_Isolation"].values
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

print("Fitted equation: y = {:.3f} * (1/x) + {:.3f}".format(model.coef_[0], model.intercept_))
print("R^2:", r2_score(y, y_pred))
```
